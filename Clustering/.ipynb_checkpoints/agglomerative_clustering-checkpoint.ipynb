```python
#kmeans with merged data from the cities
import pandas as pd
import glob, os    
import numpy as np
from sklearn.cluster import AgglomerativeClustering
from sklearn import preprocessing
import matplotlib as mpl
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as shc
from numpy import asarray
from numpy import savetxt


frankfurt =pd.read_csv('/home/madnerd/TheMiners_github/TheMiners/Clustering/Data/Frankfurt2017Prozente.csv', encoding= 'unicode_escape', sep=',', header=0, decimal = ",")
dresden = pd.read_csv('/home/madnerd/TheMiners_github/TheMiners/Clustering/Data/Dresden2017Prozente.csv', encoding= 'unicode_escape', sep=',', header=0, decimal = ",")
munchen = pd.read_csv('/home/madnerd/TheMiners_github/TheMiners/Clustering/Data/Munchen2017Prozente.csv', encoding= 'unicode_escape', sep=',', header=0, decimal = ",")
hamburg = pd.read_csv('/home/madnerd/TheMiners_github/TheMiners/Clustering/Data/Hamburg2017Prozente.csv', encoding= 'unicode_escape', sep=',', header=0, decimal = ",")
```

```python
#removing % from dresden
cols_to_check = munchen.columns[2:8]
munchen = munchen[cols_to_check].replace({'%':''}, regex=True)
munchen = munchen[cols_to_check].replace({',':'.'}, regex=True)

#removing name stadteil - replace name with number
dresden['Stadtteilname'] = dresden['Stadtteilname'].str[:2]
dresden = dresden.astype({"Stadtteilname": float})


```

```python
#path = '/home/madnerd/TheMiners_github/TheMiners/clustering_merging_data/'
#all_files = glob.glob(os.path.join(path, "*.csv"))
#df_from_each_file = (pd.read_csv(f) for f in all_files)
#concatenated_df   = pd.concat(df_from_each_file, ignore_index=True)
#concatenated_df = np.array(concatenated_df)

frames = [munchen,dresden,hamburg]
data_numbers = pd.concat(frames)

#saving input preprocessd_data into file

preprocessd_data = preprocessing.StandardScaler().fit(data_numbers).transform(data_numbers.astype(float))

data_numbers = np.array(preprocessd_data)
data_input = np.where(np.isnan(data_numbers), 0, data_numbers)

savetxt('data.csv', data_input, delimiter=',')


print(data_input)

#concatenated_df[:] = np.nan_to_num(concatenated_df)
#concatenated_df = np.array(concatenated_df)
#x = concatenated_df.iloc[:, [1,2,3,4,5,6,7,8]].values
#print(x)
```

```python
clustering = AgglomerativeClustering()
cluster_data = clustering.fit(data_input)
```

```python
plt.figure(figsize=(30, 30))
plt.title("Stadtteil Dendrogram")
dend = shc.dendrogram(shc.linkage(data_input, method='ward'))
print(dend)
```
